1)	Você fará o papel de detetive agora, e seu papel é desvendar a motivação de um crime!

Em um determinado caso de assassinato, foi deixada para trás uma pista com o motivo de tal ato. A pista é um texto misterioso que ninguém consegue decifrar. E você, como especialista em Python foi convocado para esta missão.

A pista é o texto:  **Kg8vlk8jIU7D8nNbd0GG4F2aCa4gnb85oÇAcm9aBmc55mnbtBl39Cvbn3**.

Os peritos criminais fizeram uma investigação prévia antes da sua chegada e eles trouxeram algumas dicas. Porém, não conseguiram decifrar a motivação do crime.
As dicas são:

•	A primeira letra da motivação parece estar escondida no quarto caractere da pista.

•	As outras letras estão dispersas de 5 em 5, no sentido de leitura (da esquerda para direita)

•	A motivação contém 8 caracteres.

•	Não há uma padronização de maiúsculo e minúsculo na palavra da motivação.

Com estas dicas, você, como especialista em Python com foco em resolução de crimes através de manipulação de strings, consegue resolver!

Qual foi a motivação do crime?


```python
texto = "Kg8vlk8jIU7D8nNbd0GG4F2aCa4gnb85oÇAcm9aBmc55mnbtBl39Cvbn3"
primeira_letra = texto[3]
motivacao = []
motivacao.append(primeira_letra)

for i, letra in enumerate(texto[3:]):
  if (i % 5 == 0 and i != 0):
    motivacao.append(letra.lower())

print(motivacao[:8])
```

2)	Uma empresa de varejo quer analisar as vendas históricas de seus produtos para tomar decisões estratégicas.

image.png

A - As informações de vendas diárias são mutáveis ou imutáveis? A estrutura de dados usada é a mais adequada? Justifique sua resposta.

B - As informações específicas de cada produto são mutáveis ou imutáveis? A estrutura de dados usada é a mais adequada? Justifique sua resposta.

C - Caso as estruturas sejam inadequadas, como você sugeriria?

```python
vendas_diarias = (
    ["Produto A", 10, 100],
    ["Produto B", 5, 250],
    ["Produto C", 2, 150],
)

print("A estrutura é uma tupla e as tuplas são imutáveis, as informações específicas também são imutáveis, logo não podem ser alteradas.")
print("Neste caso, minha sugestão seria trabalhar com dicionários que são mais fáceis de acessar, manipular e trabalhar.")
print("Tuplas são interessantes quando os dados setados em sua declaração não podem ser alterados e manipulados.")

vendas_diarias = {
    "Produto A" : { "Quantidade": 10, "Preco": 100 },
    "Produto B" : { "Quantidade": 5, "Preco": 250 },
    "Produto C" : { "Quantidade": 2, "Preco": 150 },
}

total_vendas = sum(vendas_diarias[produto]['Quantidade'] * vendas_diarias[produto]['Preco'] for produto in vendas_diarias.keys())
print('Faturamento: R$', total_vendas)
```

4) Uma concessionária de automóveis precisa gerenciar o estoque de peças de reposição em seus diferentes depósitos. O gerente gostaria de ter uma visão sobre o gerenciamento de controle de estoque das peças disponíveis para manutenção.
Abaixo há uma lista de todos os depósitos com suas respectivas peças em estoque:

*Depósito A: junta homocinética, filtro de óleo, vela de ignição, correia dentada, pastilha de freio, amortecedor*

*Depósito B: filtro de óleo, vela de ignição, pastilha de freio, disco de freio, pneu, válvula termostática*

*Depósito C: correia dentada, pastilha de freio, amortecedor, bateria, filtro de óleo, alternador, vela de ignição*

*Depósito D: vela de ignição, disco de freio, óleo de motor, correia alternador, sensor de oxigênio, fluido de freio*

*Depósito E: pneu, amortecedor, radiador, bomba d’água, mangueira de água, rolamento de roda, filtro de óleo, vela de ignição*

Responda as seguintes perguntas:

1 - Quais peças estão disponíveis em todos os depósitos?

2 - Quais peças que estão faltando em cada um dos depósitos?

3	- Quais peças são exclusivas de cada depósito?


```python
estoque = {
    'Depósito A': ['junta homocinética', 'filtro de óleo', 'vela de ignição', 'correia dentada', 'pastilha de freio', 'amortecedor'],
    'Depósito B': ['filtro de óleo', 'vela de ignição', 'pastilha de freio', 'disco de freio', 'pneu', 'válvula termostática'],
    'Depósito C': ['correia dentada', 'pastilha de freio', 'amortecedor', 'bateria', 'filtro de óleo', 'alternador', 'vela de ignição'],
    'Depósito D': ['vela de ignição', 'disco de freio', 'óleo de motor', 'correia alternador', 'sensor de oxigênio', 'fluido de freio'],
    'Depósito E': ['pneu', 'amortecedor', 'radiador', 'bomba d’água', 'mangueira de água', 'rolamento de roda', 'filtro de óleo', 'vela de ignição'],
}

# -------------------

lista_completa = [ item for deposito in estoque.values() for item in deposito ]
pecas_unicas = list(set(lista_completa))

# ou

pecas_unicas = set()
for itens in estoque.values():
    pecas_unicas.update(itens)
print('Lista completa de peças de todos os estoques:', pecas_unicas)

# -------------------

print('\n')

for deposito, itens in estoque.items():
    presentes = set(itens)
    faltando = pecas_unicas - presentes
    print(f"Itens faltando no {deposito}:", faltando)

# -------------------

print('\n')

conjuntos = [set(itens) for itens in estoque.values()]
comum = set.intersection(*conjuntos)

print("Peças presentes em todos os depósitos:")
for peca in comum:
    print("-", peca)

# -------------------

print('\n')

for deposito, itens in estoque.items():
    outros = set()
    for k, v in estoque.items():
        if k != deposito:
            outros.update(v)
    exclusivas = set(itens) - outros
    print(f"Peças exclusivas do {deposito}: {list(exclusivas) if exclusivas else 'Nenhuma'}")
```

5) A tabela a seguir contém intervalos de magnitude de terremotos na escala Richter e a descrição do impacto de cada intervalo:

image.png

Escreva um programa que leia uma magnitude fornecida pelo usuário e exiba o descritor apropriado como parte de uma mensagem significativa.
Por exemplo, se o usuário inserir 5.5, seu programa deve indicar que um terremoto de magnitude 5.5 é considerado um terremoto moderado.

```python
magnitude = float(input("Entre com a magnitude (por ex: 5.5): "))

if magnitude < 2.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto micro")
elif magnitude >= 2.0 and magnitude < 3.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto muito pequeno")
elif magnitude >= 3.0 and magnitude < 4.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto pequeno")
elif magnitude >= 5.0 and magnitude < 6.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto leve")
elif magnitude >= 6.0 and magnitude < 7.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto moderado")
elif magnitude >= 7.0 and magnitude < 8.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto forte")
elif magnitude >= 8.0 and magnitude < 9.0:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto grande")
elif magnitude >= 9.0 and magnitude < 10:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto enorme")
elif magnitude >= 10:
  print(f"Um terremoto de magnitude {magnitude} é considerado um terremoto catastrofico")
else:
  print(f"Não mapeado")
```

6) Um plano de celular específico inclui 50 minutos de tempo de ligação e 50 mensagens de texto por 100,00 reais por mês. Cada minuto adicional de tempo de ligação custa 0,25, enquanto mensagens de texto adicionais custam 0,15 cada.

Todas as contas de celular incluem uma taxa adicional de 2,00 para suportar os centros de atendimento de emergência 191, e o valor total da conta (incluindo a taxa do 911) está sujeito a um imposto de 5%.

Escreva um programa que leia do usuário o número de minutos e mensagens de texto utilizados em um mês. Exiba o valor base, a cobrança por minutos adicionais (se houver), a cobrança por mensagens de texto adicionais (se houver), a taxa do 191, o imposto e o valor total da conta.

Exiba as cobranças por minutos e mensagens de texto adicionais apenas se o usuário tiver incorrido em custos nessas categorias. Certifique-se de que todos os valores sejam exibidos com 2 casas decimais.


```python
plano = { 'minutos': 50, 'mensagens': 50, 'valor': 100, 'minuto_adicional': 0.25, 'mensagem_adicional': 0.15, 'taxa': 2.00, 'imposto': 5 }

minutos = float(input("Quantos minutos foram utilizados? "))
mensagens = int(input("Quantas mensagens foram utilizadas? "))

print('Valor base do plano: R$', '%.2f' % round(float(plano["valor"])))
print('Minutos contratado:', plano["minutos"])
print('Mensagens contratada:', plano["minutos"])

if minutos > plano["minutos"]:
  minutos_add = minutos - plano["minutos"]
  valor_minutos_add = round(minutos_add * plano['minuto_adicional'], 2)
  print('Total de minutos utilizado:', '%.2f' % minutos, '- Valor adicional a ser cobrado a: R$', '%.2f' % valor_minutos_add, '- Minutos adicionais: ', round(minutos_add, 2))

if mensagens > plano["mensagens"]:
  mensagens_add = mensagens - plano["mensagens"]
  valor_mensagens_add = round(mensagens_add * plano['mensagem_adicional'], 2)
  print('Total de mensagens utilizado:', mensagens, '- Valor adicional a ser cobrado: R$', '%.2f' % valor_mensagens_add, '- Mensagens adicionais: ', round(mensagens_add, 2))

print('Taxa adicional: R$', '%.2f' % plano['taxa'])

valor_conta = float(plano["valor"]) + valor_minutos_add + valor_mensagens_add + plano['taxa']

imposto = valor_conta * plano['imposto'] / 100

print(f"Valor do imposto: R$ {'%.2f' % round(imposto, 2)} - Imposto de {plano['imposto']}%")
print(f"Valor a pagar: R$ {'%.2f' % round(valor_conta + imposto, 2)}")
```
